```python
import pandas as pd
data=pd.read_csv('dataset.csv')
print(data)
```

```python
X=data.X1
y=data.y
print(X)
print(y)
```

```python


X_mean=X.mean()
ymean=y.mean()
numinator=(X-X_mean) *( y-ymean)
numinator=numinator.sum()
denominator=(X-X_mean)**2
denominator=denominator.sum()
m=numinator/denominator
c=ymean-(m*X_mean) 
print(m)
print(c)
```

```python
#plot linearregression
import matplotlib.pyplot as plt
plt.scatter(X,y)
plt.plot([min(X),max(X)],[m*min(X)+c,m*max(X)+c],color='red')
plt.show()
```

```python
ypredicted=m*X+c
meansquareerror=(y-ypredicted)**2
meansquareerror=meansquareerror.sum()
meansquareerror=meansquareerror/len(X)
print(meansquareerror)
```
